```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
from plotter import Plotter
%matplotlib inline
```

```python
## set the dir of the results
train_file = './results/train_e_'
test_file = './results/test_e_'
## set the save path
path = './figures/ResNet_exits.png'

train = 0
test = 0

for i in range(20):
    train += genfromtxt(train_file + str(i + 1) + '.csv', delimiter = ',')
    test += genfromtxt(test_file + str(i + 1) + '.csv', delimiter = ',')

train = (train / 20).T
test = (test / 20).T

#train = genfromtxt(train_file, delimiter = ',').T
#test = genfromtxt(test_file, delimiter = ',').T
```

```python
# split the data into shape (number of exits, epoches)
train_loss = train[[0, 2, 4], :]
train_acc = train[[1, 3, 5], :]
test_loss = test[[0, 2, 4], :]
test_acc = test[[1, 3, 5], :]
```

```python
plotter = Plotter(title="ResNet with three exits",
                         save_to_filepath=path)
```

```python
plotter.plot_values(train_loss, test_loss, train_acc, test_acc)
```
